```python
import nibabel as nib
import numpy as np
import os
import glob
```

```python
npys = glob.glob('*.npy')
```

```python
npys
```

```python
np.load(npys[0]).shape
```

```python
ref_bold_nii = nib.load('sub-NeuroMET00401_task-rest_space-MNI152NLin2009cAsym_desc-brain_mask.nii.gz')
```

```python
template = 'tpl-MNI152NLin2009cAsym_res-01_T1w.nii.gz'
```

```python
template_nii = nib.load(template)
```

```python
template_nii.shape
```

```python
from nilearn.image import resample_img
for npy in npys:
    data = np.load(npy)
    # Reshape the data with nilearn if necessary
    if data.shape != template_nii.shape:
        data = resample_img(nib.Nifti1Image(data, ref_bold_nii.affine, ref_bold_nii.header), target_shape=template_nii.shape, target_affine=template_nii.affine).get_fdata()

    new_nii = nib.Nifti1Image(data, template_nii.affine, template_nii.header)
    outname = npy.replace('.npy', '.nii.gz')
    nib.save(new_nii, outname)
    print(f'Saved {outname}')
```
